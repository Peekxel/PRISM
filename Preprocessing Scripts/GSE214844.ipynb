```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
adata = sc.read_h5ad('GSE214844_antNPC.MNN.KOvsWT.h5ad')
adata_day5_cm = sc.read_h5ad('GSE214844_antNPC.MNN.KOvsCMvsWT.h5ad')
adata_treat = sc.read_h5ad('GSE214844_antNPC.MNN.treatment.h5ad')
adata_treat
```

```python
(
    adata_treat.obs['treatment'].value_counts(),
    adata_day5_cm.obs['treatment'].value_counts(),
    adata.obs['treatment'].value_counts(),
)
```

```python
(
    adata_treat.obs['old.ident'].value_counts(),
    adata_day5_cm.obs['old.ident'].value_counts(),
    adata.obs['old.ident'].value_counts()
)
```

```python
(
    adata.obs_names.isin(adata_treat.obs_names).sum(),
    adata.obs_names.isin(adata_day5_cm.obs_names).sum()
)
```

```python
ko_wt_day3_indices = adata[adata.obs['old.ident'].isin(['WT_day3', 'KO_day3'])].obs_names
ko_wt_day5_indices = adata[adata.obs['old.ident'].isin(['WT_day5', 'KO_day5'])].obs_names
```

```python
adata_merged = ad.concat([
    adata[ko_wt_day3_indices],
    adata_day5_cm,
    adata_treat[~adata_treat.obs_names.isin(ko_wt_day5_indices)]
])
adata_merged
```

```python
COLS, ORGANISMS
```

```python
adata.obs.head()
```

```python
adata_merged.obs['condition'] = adata_merged.obs['genotype']
```

```python
adata_merged.obs['perturbation_name'] = adata_merged.obs['condition'].apply(lambda x: 'TET1' if x != 'WT' else 'non-targeting')
adata_merged.obs['crispr_type'] = 'CRISPR KO'
adata_merged.obs['cancer_type'] = 'non-cancer'
adata_merged.obs['cell_type'] = 'Embryonic stem cells'
adata_merged.obs['organism'] = ORGANISMS[1]
```

```python
missing = [col for col in COLS if col not in adata_merged.obs]
print(f"Missing following columns: \n {missing}")

if 'organism' in adata.obs_keys():
    assert adata_merged.obs['organism'].isin(ORGANISMS).all(), "Invalid organism naming"
```

```python
adata_merged.write_h5ad('GSE214844_merged.h5ad')
```
