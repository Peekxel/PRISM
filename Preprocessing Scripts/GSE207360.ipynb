```python
import re

expression_data = {}
metadata = {}

with tarfile.open('GSE207360_RAW.tar') as tar:
    for member in tar.getmembers():
        if 'dcc' in member.name:
            print(f"Processing: {member.name}")
            gz_file = tar.extractfile(member)
            with gzip.open(gz_file) as f:
                content = f.read().decode('utf-8')

                sample_metadata = {}
                for section in ['<Header>', '<Scan_Attributes>', '<NGS_Processing_Attributes>']:
                    section_match = re.search(f'{section}(.*?)</', content, re.DOTALL)
                    if section_match:
                        section_content = section_match.group(1).strip().split('\n')
                        for line in section_content:
                            if ',' in line:
                                key, value = line.split(',', 1)
                                sample_metadata[key] = value.strip('"')

                code_summary_match = re.search('<Code_Summary>(.*?)</Code_Summary>', content, re.DOTALL)
                if code_summary_match:
                    code_summary = code_summary_match.group(1).strip().split('\n')
                    sample_data = {line.split(',')[0]: int(line.split(',')[1]) for line in code_summary}

                gsm_id = member.name.split('_')[0]
                roi_id = sample_info[gsm_id]['sample']
                expression_data[roi_id] = sample_data
                metadata[roi_id] = {
                    'GSM_ID': gsm_id,
                    'ROI_ID': roi_id,
                    'condition': sample_info[gsm_id]['condition'],
                    **sample_metadata
                }

expression_matrix = pd.DataFrame(expression_data).fillna(0)

metadata_df = pd.DataFrame.from_dict(metadata, orient='index')

expression_matrix.head()
```

```python
import json

with open('GSE207360_MMR_R_NGS_WTA_v1.0.pkc', 'rb') as f:
    pkc_data = json.load(f)
```

```python
rts_to_gene = {}
for target in pkc_data['Targets']:
    for probe in target['Probes']:
        rts_id = probe['RTS_ID']
        gene_name = target['DisplayName']
        rts_to_gene[rts_id] = gene_name
```

```python
expression_matrix.index = expression_matrix.index.map(lambda x: rts_to_gene.get(x, x))
adata = ad.AnnData(X=expression_matrix.T, obs=metadata_df)
adata.var_names = expression_matrix.index
```

```python
adata.obs['perturbation_name'] = adata.obs['condition'].apply(lambda x: 'EGFR' if x == 'KO' else x)
adata.obs['crispr_type'] = 'CRISPR KO'
adata.obs['cancer_type'] = 'Glioblastoma'
adata.obs['cell_type'] = 'CD31+ blood vessels'
adata.obs['organism'] = ORGANISMS[1]
```
