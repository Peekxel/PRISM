```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
def assign_guides(adata, guide_df):
    for i, guide in enumerate(guide_df.iloc[:, 0]):
        adata.obs[guide] = adata.obs_names.isin(
            guide_df.iloc[i, 1].split(', ')
        )

    adata.obs['guides'] = adata.obs[guide_df[0]].apply(
        lambda x: "+".join(guide_df[0].values[x.values]), axis=1
    )
    adata.obs["perturbation_name"] = adata.obs["guides"].apply(
        lambda x: "+".join(list(set([i.split('_')[1].lstrip('sg') for i in x.split('+') if len(i) > 0])))
    )
    adata.obs.drop(columns=guide_df[0].values, inplace=True)
```

```python
samples = ["_".join(file.split('.')[0].split("_")[:]) for file in os.listdir() if file.endswith('.mtx.txt.gz')]
guide_infos = {}
for sample in samples:
    guide_infos[sample] = []
    for file in os.listdir():
        if sample in file and 'cbc_gbc_dict' in file:
            guide_infos[sample].append(file)
adatas_mouse = {}
adatas_human = {}

for sample in samples:
    adata = sc.read_mtx(f"{sample}.mtx.txt.gz").T
    adata.obs = pd.read_csv(f"{sample}_cellnames.csv.gz", index_col=0)
    adata.var = pd.read_csv(f"{sample}_genenames.csv.gz", index_col=0)
    adata.var["gene_id"] = adata.var['0'].str.split('_').str[0]
    adata.var["gene_name"] = adata.var['0'].str.split('_').str[1]
    adata.obs.set_index("0", inplace=True)
    adata.var.set_index("gene_name", inplace=True)
    guide_df = pd.read_csv(guide_infos[sample][0], header=None)
    assign_guides(adata, guide_df)
    adata.obs_names_make_unique()
    adata.var_names_make_unique()
    if 'k562' in sample:
        adatas_human[sample] = adata
        adata.obs['organism'] = ORGANISMS[0]
        adata.obs['cancer_type'] = 'Erythroleukemia'
        adata.obs['cell_type'] = 'erythroleukemia cell'
        adata.obs['crispr_type'] = 'CRISPR KO'
    else:
        adatas_mouse[sample] = adata
        adata.obs['organism'] = ORGANISMS[1]
        adata.obs['cancer_type'] = 'non-cancer'
        adata.obs['cell_type'] = 'embryonic stem cellsl'
        adata.obs['crispr_type'] = 'CRISPR KO'

adata_mouse = ad.concat(adatas_mouse, label="sample")
adata_human = ad.concat(adatas_human, label="sample")
```

```python
adata_human.obs['condition'] = adata_human.obs['perturbation_name'].apply(
    lambda x: 'non-targeting' if 'INTERGENIC' in x and len(x.split('+')) == 1 else 'targeting'
)
adata_mouse.obs['condition'] = adata_mouse.obs['perturbation_name'].apply(
    lambda x: 'non-targeting' if 'MouseNTC' in x and len(x.split('+')) == 1 else 'targeting'
)
```

```python
missing = [col for col in COLS if col not in adata_human.obs]
print(f"Missing following columns: \n {missing}")

if 'organism' in adata_human.obs_keys():
    assert adata_human.obs['organism'].isin(ORGANISMS).all(), "Invalid organism naming"

missing = [col for col in COLS if col not in adata_mouse.obs]
print(f"Missing following columns: \n {missing}")

if 'organism' in adata_mouse.obs_keys():
    assert adata_mouse.obs['organism'].isin(ORGANISMS).all(), "Invalid organism naming"
```
