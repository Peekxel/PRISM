```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
adatas = []
for file in os.listdir('.'):
    if file.endswith('.h5') and file != 'GSE215253_mouse_transplant_filtered_feature_bc_matrix.h5':
        adata = sc.read_10x_h5(file)
        if 'perturb_BE' in file:
            perturbation = file.split('_')[4]
            condition = 'perturb_BE_seq'
        else:
            perturbation = file.split('_')[3]
            condition = 'perturb_seq'
        adata.obs['perturbation_name'] = perturbation
        adata.obs['condition'] = condition
        adata.obs['replicate'] = file.split('_')[5]
        adata.var_names_make_unique()
        adatas.append(adata)
adata = ad.concat(adatas)
adata
```

```python
adata.obs['crispr_type'] = adata.obs['condition'].apply(
    lambda x: 'CRISPR Base Editing' if x == 'perturb_BE_seq' else 'CRISPR KO'
)
adata.obs['cancer_type'] = 'non-cancer'
adata.obs['cell_type'] = 'hematopoietic stem and progenitor cell'
adata.obs['organism'] = ORGANISMS[0]
```

```python
adata.obs.rename(columns={'perturbation': 'perturbation_name'}, inplace=True)
```

```python
adata.write_h5ad('GSE215253_merged.h5ad')
```

```python
missing = [col for col in COLS if col not in adata.obs]
print(f"Missing following columns: \n {missing}")

if 'organism' in adata.obs_keys():
    assert adata.obs['organism'].isin(ORGANISMS).all(), "Invalid organism naming"
```
