```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
normed_counts = pd.read_csv(
    'GSE149209_Ensembl.exon.rpkm.tsv.gz',
    sep='\t',
    index_col=0,
)
counts = pd.read_csv(
    'GSE149209_Ensembl.exon.raw.tsv.gz',
    sep='\t',
    index_col=0,
)
normed_counts.head()
```

```python
adata = ad.AnnData(
    X=normed_counts.values.T,
    obs=pd.DataFrame(index=normed_counts.columns),
    var=pd.DataFrame(
        index=normed_counts.index.str.split(':').str[1],
        data={'gene_id': normed_counts.index.str.split(':').str[0]},
    ),
    layers={'counts': counts.values.T},
)
adata
```

```python
adata.obs['edit_type'] = adata.obs.index.str.split('__').str[-2].str.split('-').str[-1]

crispr_type_map = {
    'unedited': 'None',
    'parental': 'None',
    'KO': 'CRISPR KO',
    'repair': 'CRISPR KO'
}
adata.obs['crispr_type'] = adata.obs['edit_type'].map(crispr_type_map)

condition_map = {
    'unedited': 'H3.3G34W_mutant',
    'parental': 'control',
    'KO': 'H3.3G34W_edited',
    'repair': 'H3.3G34W_repaired'
}
adata.obs['condition'] = adata.obs['edit_type'].map(condition_map)

perturbation_map = {
    'unedited': 'non-targeting',
    'parental': 'non-targeting',
    'KO': 'H3F3A',
    'repair': 'H3F3A'
}
adata.obs['perturbation_name'] = adata.obs['edit_type'].map(perturbation_map)
```

```python
adata.obs['cell_type'] = 'GCT stromal cell'
adata.obs['cancer_type'] = 'Giant Cell Tumor of Bone'
adata.obs['organism'] = ORGANISMS[0]
```
