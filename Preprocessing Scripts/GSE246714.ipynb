```python
adatas = {}
genes = pd.read_csv('GSE246714_b_all_genes.tsv.gz', sep='\t')
for sample in samples:
    adata = sc.read_mtx(f"{sample}_matrix.mtx.gz").T
    barcodes = pd.read_csv(f"{sample}_barcodes.tsv.gz", header=None)
    adata.obs.index = barcodes.iloc[:, 0]
    adata.obs.index.name = 'cell_barcode'
    adata.obs_names_make_unique()
    adatas[sample] = adata

adata = ad.concat(adatas, label='sample')
adata.var.index = genes.iloc[:, 0]
adata.write_h5ad('GSE246714_merged.h5ad')
```
