```r
seurat.data <- readRDS("./GSE150062_seurat.rds")
seurat.data
```

```r
head(seurat.data@meta.data)
```

```r
cols <- c('guide_identity', 'guide_target', 'condition')
message(unique(seurat.data$condition))
head(seurat.data@meta.data[, cols])
```

```r
obs <- seurat.data@meta.data
obs$cell_type <- 'iPSCs (induced pluripotent stem cells)'
obs$organism <- "Humans (Homo sapiens)"
obs$crispr_type <- "CRISPRi"
obs$cancer_type <- "non-cancer"
obs$condition <- "Neural differentiation"
names(obs)[names(obs) == 'guide_target'] <- "perturbation_name"
names(obs)[names(obs) == 'guide_identity'] <- "guide"
head(obs)
```

```r
write.csv(obs, "GSE150062_obs.csv")
```
