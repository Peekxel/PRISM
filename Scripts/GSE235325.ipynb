```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
adata = sc.read_10x_h5("GSE235325_P4.wta4to10_filtered_seurat_merged_counts.h5")
adata
```

```python
import re
import numpy as np

obs = pd.DataFrame()
with tarfile.open('GSE235325_RAW.tar') as tar:
    for member in tar.getmembers():
        sample = member.name.split('_')[3]
        num = re.findall('(WTA)(\d+)', sample)[0][1]
        if int(num) > 10:
            break
        file = tar.extractfile(member)
        with gzip.open(file) as gz_file:
            sgrna = pd.read_csv(gz_file)
            sgrna.index = sample.lower() + '_' + sgrna['id'].astype(str)
            obs = pd.concat([obs, sgrna])

indicies = np.intersect1d(adata.obs_names, obs.index)

obs = obs.loc[indicies]
adata = adata[indicies]
adata.obs = obs
```

```python
import re

obs = pd.DataFrame()
with tarfile.open('GSE235325_RAW.tar') as tar:
    for member in tar.getmembers():
        sample = member.name.split('_')[3]
        num = re.findall('(WTA)(\d+)', sample)[0][1]
        if int(num) < 11:
            continue
        file = tar.extractfile(member)
        with gzip.open(file) as gz_file:
            sgrna = pd.read_csv(gz_file)
            sgrna.index = sample.lower() + '_' + sgrna['id'].astype(str)
            obs = pd.concat([obs, sgrna])
```

```python
adata.obs.index = adata.obs_names.to_series().apply(
    lambda x: "_".join([y if i == 0 else str(int(float(y))) for i, y in enumerate(x.split('_'))]) if 'e' in x else x
)
```

```python
indicies = np.intersect1d(adata.obs_names, obs.index)

obs = obs.loc[indicies]
adata = adata[indicies]
adata.obs = obs
adata
adata.write_h5ad('GSE235325_wta11to18.h5ad')
```

```python
adata0 = sc.read_h5ad('GSE235325_wta4to10.h5ad')
adata1 = sc.read_h5ad('GSE235325_wta11to18.h5ad')
adata = ad.concat([adata0, adata1], join='outer')
adata
```

```python
adata.obs['crispr_type'] = 'CRISPR KO'
adata.obs['cancer_type'] = 'skin squamous cell carcinoma'
adata.obs['cell_type'] = 'skin cell'
adata.obs['organism'] = ORGANISMS[1]
```

```python
adata.obs['perturbation_name'] = adata.obs['sgRNA'].str.split('_').str[0].apply(
    lambda x: x if x != 'ctrl' else 'non-targeting'
)
adata.obs['condition'] = adata.obs['sgRNA'].str.split('_').str[0].apply(
    lambda x: 'targeting' if x != 'ctrl' else 'non-targeting'
)
```
