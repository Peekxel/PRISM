```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

## scRNA-seq

```python
samples = [sample for sample in os.listdir() if os.path.isdir(sample)]
```

```python
adatas = []

for sample in samples:
    adata = sc.read_10x_mtx(sample)
    adata.obs['sample'] = sample
    adatas.append(adata)
    break

adata = ad.concat(adatas)
adata.write_h5ad('GSE196142_merged.h5ad')
```

```python
adata = sc.read_h5ad('GSE196142_merged.h5ad')
adata.var.head()
```

```python
adata.obs['crispr_type'] = 'CRISPR KO'
adata.obs['cancer_type'] = 'Leukemia'
adata.obs['cell_type'] = 'B-cell acute lymphoblastic leukemia (B-ALL) cell'
adata.obs['organism'] = ORGANISMS[1]
adata.obs['condition'] = 'control'
adata.obs['perturbation_name'] = 'non-targeting'
```

```python
adata.write_h5ad('GSE196142_merged.h5ad')
```

```python
bulk_samples = [file for file in os.listdir('mageck') if 'all_v_input.normalized' in file]
bulk_samples
```

```python
for i, sample in enumerate(bulk_samples):
    if i == 0:
        counts = pd.read_csv(f'mageck/{sample}', sep='\t', index_col=0)
    else:
        counts = pd.concat([counts, pd.read_csv(f'mageck/{sample}', sep='\t', index_col=0)], axis=0)
```

```python
adata = ad.AnnData(counts.iloc[:, 1:].T,)
adata.var['sgRNA'] = adata.var.index
adata.var['perturbation_name'] = counts['Gene']
adata.var.index = counts['Gene']
adata
```

```python
adata.var['perturbation_name'] = adata.var['perturbation_name'].apply(
    lambda x: 'non-targeting' if 'NO_SITE' in x or 'INTERGENIC' in x else x
)
adata.var['perturbation_name'] = adata.var['perturbation_name'].apply(
    lambda x: '+'.join(x.split('_'))
)
adata.var['perturbation_name']
```

```python
adata.var['condition'] = adata.var['perturbation_name'].apply(
    lambda x: x if x == 'non-targeting' else 'targeting'
)
adata.var['condition']
```

```python
adata.obs['crispr_type'] = 'CRISPR KO'
adata.obs['cancer_type'] = 'Leukemia'
adata.obs['cell_type'] = 'B-cell acute lymphoblastic leukemia (B-ALL) cell'
adata.obs['organism'] = ORGANISMS[1]
```

```python
adata.write_h5ad('GSE196142_bulk_merged.h5ad')
```
