```python
import numpy as np


def read_10x_mtx_legacy(prefix):
    adata = sc.read_mtx(f"{prefix}matrix.mtx.gz")
    barcodes = pd.read_csv(f"{prefix}cells.csv.gz", header=None,)
    barcodes.columns = ['cell_barcode']
    adata.obs.index = barcodes['cell_barcode']
    genes = pd.read_csv(f"{prefix}genes.csv.gz", header=None, index_col=0)
    genes.index.name = 'gene_name'
    adata.var = genes
    return adata

def read_10x_mtx_processed(prefix):
    sample = prefix.split('.')[0]
    adata = read_10x_mtx_legacy(f"{sample}.rawcounts.")
    barcodes = pd.read_csv(f"{prefix}cells.csv.gz", header=None,)
    barcodes.columns = ['cell_barcode']
    adata = adata[barcodes['cell_barcode']]
    adata.raw = adata

    adata.obs.index = barcodes['cell_barcode']
    genes = pd.read_csv(f"{prefix}genes.csv.gz", header=None, index_col=0)
    genes.index.name = 'gene_name'
    adata = adata[:, genes.index]
    adata.var = genes

    adata.layers['zscored'] = sc.read_mtx(f"{prefix}matrix.mtx.gz").X

    cell_meta = pd.read_csv(f'{prefix}cells.metadata.csv.gz', index_col='cell')
    adata.obs = cell_meta.loc[:, (~cell_meta.columns.str.startswith('PC')) & (~cell_meta.columns.str.startswith('UMAP'))]
    adata.obsm['X_pca'] = cell_meta.loc[:, cell_meta.columns.str.startswith('PC')].values
    adata.obsm['X_umap'] = cell_meta.loc[:, cell_meta.columns.str.startswith('UMAP')].values

    gene_meta = pd.read_csv(f'{prefix}genes.metadata.csv.gz', index_col=0)
    adata.var = gene_meta.loc[:, ~gene_meta.columns.str.startswith('PC')]
    adata.varm['X_pca'] = gene_meta.loc[:, gene_meta.columns.str.startswith('PC')].values

    sgrna_assignment = pd.read_csv(f"{prefix.split('.')[0]}.variants2cell.csv.gz", index_col='cell', sep='\t')
    index = np.intersect1d(sgrna_assignment.index, adata.obs_names)
    adata = adata[index]
    sgrna_assignment = sgrna_assignment.loc[index]

    adata.obs.loc[index, sgrna_assignment.columns[-2:]] = sgrna_assignment.iloc[:, -2:]
    adata.obs['perturbation_name'] = adata.obs['variant.detailed_multi'].apply(
        lambda x: '+'.join(x.split(','))
    )
    adata.obs['perturbation_name'] = adata.obs['perturbation_name'].apply(
        lambda x: 'non-targeting' if x == 'unassigned' else x
    )

    return adata
```

```python
adata_kras = read_10x_mtx_processed('GSE161824_A549_KRAS.processed.')
adata_tp53 = read_10x_mtx_processed('GSE161824_A549_TP53.processed.')
adata = adata_kras.concatenate(adata_tp53, batch_key='variant', batch_categories=['KRAS', 'TP53'])
adata, adata.raw.X.shape
```

```python
adata.obs['condition'] = adata.obs['variant'].astype(str)
adata.obs.loc[adata.obs['perturbation_name'] == 'non-targeting', 'condition'] = 'control'
```
