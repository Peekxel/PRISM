```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
samples = ["_".join(file.split(".")[0].split("_")[:-1]) for file in os.listdir() if file.endswith("matrix.mtx.gz")]
samples
```

```python
def read_10x_mtx_legacy(prefix):
    adata = sc.read_mtx(f"{prefix}_matrix.mtx.gz").T
    barcodes = pd.read_csv(f"{prefix}_barcodes.tsv.gz", header=None, sep='\t')
    barcodes.columns = ['cell_barcode']
    adata.obs.index = barcodes['cell_barcode']
    genes = pd.read_csv(f"{prefix}_genes.tsv.gz", header=None, sep='\t', index_col=1)
    genes.index.name = 'gene_name'
    genes.columns = ['gene_id']
    adata.var = genes
    return adata
```

```python
adatas = []
for sample in samples:
    print(sample)
    adata = read_10x_mtx_legacy(sample)
    barcodes = pd.read_csv(f"{sample}_filtered_barcodes.tsv.gz", sep='\t', header=None, index_col=0)
    adata.var_names_make_unique()
    adata = adata[barcodes.index.values]
    adata.obs['sample'] = sample
    adatas.append(adata)

adata = ad.concat(adatas)
adata.write("expr_raw.h5ad")
```
