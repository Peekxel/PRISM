```python
%pip install scanpy python-dotenv -q

import os
import tarfile
import json
import gzip

import scanpy as sc
import anndata as ad
import pandas as pd

from dotenv import load_dotenv
load_dotenv(dotenv_path='/content/drive/MyDrive/.gse.env')
COLS = json.loads(os.environ.get('OBS_COLS'))
ORGANISMS = json.loads(os.environ.get('ORGANISMS'))
```

```python
adatas = {}
for file in [
    "GSM8282886_OE1_filtered_feature_bc_matrix.h5",
    "GSM8282888_NT1_filtered_feature_bc_matrix.h5",
    "GSM8282890_OE2_filtered_feature_bc_matrix.h5",
    "GSM8282892_NT2_filtered_feature_bc_matrix.h5",
]:
    adata = sc.read_10x_h5(file)
    adata.var_names_make_unique()
    adata.obs_names_make_unique()
    adatas['_'.join(file.split("_")[:2])] = adata

adata = ad.concat(adatas, label='sample')
adata.write_h5ad("GSE267982_merged.h5ad")
```
