```python
def get_perturbation(ident):
    if ident.startswith('PGP'):
        return 'non-targeting'
    elif ident.startswith('D1') or ident.startswith('M2'):
        return 'HNRNPU'

def get_condition(ident):
    if ident.startswith('PGP'):
        return 'non-targeting'
    elif ident.startswith('D1') or ident.startswith('M2'):
        return 'targeting'

adata.obs['perturbation_name'] = adata.obs['orig.ident'].apply(get_perturbation)
adata.obs['perturbation_name']
```

```python
adata.obs['perturbation_name'].unique()
```

```python
adata.obs['condition'] = adata.obs['orig.ident'].apply(get_condition)
```

```python
adata.var_names
```

```python
adata.obs['crispr_type'] = 'CRISPR KO'
adata.obs['cancer_type'] = 'non-cancer'
adata.obs['cell_type'] = 'cortical organoid'
adata.obs['organism'] = ORGANISMS[0]
```

```python
missing = [col for col in COLS if col not in adata.obs]
print(f"Missing following columns: \n {missing}")

if 'organism' in adata.obs_keys():
    assert adata.obs['organism'].isin(ORGANISMS).all(), "Invalid organism naming"
```

## Bulk

```python
%%bash

urls=$(cat <<'EOF'
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5FE13%5FHnrnpu%2B%2D%2Ecsv%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5FP0%5FHnrnpu%2B%2D%2Ecsv%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5FP1%5FHnrnpu%5Ffl%5Fhet%2Ecsv%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5FP1%5FHnrnpu%5Ffl%5Fhom%2Ecsv%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5Fest%5Fcounts%5Fgenes%5Fkallisto%5FE13%5FHNRNPU%2B%2D%2Etxt%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5Fest%5Fcounts%5Fgenes%5Fkallisto%5FP0%5FHNRNPU%2B%2D%2Etxt%2Egz
https://www.ncbi.nlm.nih.gov/geo/download/?acc=GSE219317&format=file&file=GSE219317%5Fest%5Fcounts%5Fgenes%5Fkallisto%5FP0%5FHNRNPUfl%2Dandflfl%2Etxt%2Egz
EOF
)

urldecode() {
    local url_encoded="${1//+/ }"
    printf '%b' "${url_encoded//%/\\x}"
}

while IFS= read -r url
do
    # Extract the filename from the URL
    encoded_filename=$(echo "$url" | sed -n 's/.*file=\(.*\)/\1/p')

    # Decode the filename
    decoded_filename=$(urldecode "$encoded_filename")

    # Download the file
    curl -o "$decoded_filename" "$url"

    echo "Downloaded: $decoded_filename"
done <<< "$urls"
```
